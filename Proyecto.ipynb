```python
import pandas as pd
import re
def eliminar_emojis(texto):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # Emoticonos y emojis en caras
                               u"\U0001F300-\U0001F5FF"  # Símbolos y pictogramas
                               u"\U0001F680-\U0001F6FF"  # Transporte y símbolos de mapas
                               u"\U0001F700-\U0001F77F"  # Símbolos de alquimia
                               u"\U0001F780-\U0001F7FF"  # Símbolos de geografía
                               u"\U0001F800-\U0001F8FF"  # Símbolos de emoji extendidos
                               u"\U0001F900-\U0001F9FF"  # Símbolos de emoji extendidos
                               u"\U0001FA00-\U0001FA6F"  # Símbolos de emoji extendidos
                               u"\U0001FA70-\U0001FAFF"  # Símbolos de emoji extendidos
                               u"\U0001F004-\U0001F0CF"  # Símbolos de cartas
                               u"\U0001F170-\U0001F251"  # Símbolos de letras
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', texto)

# Aplicar la función eliminar_emojis a la columna 'Texto'
data_final['Texto_limpio'] = data_final['text'].apply(eliminar_emojis)

# Mostrar el DataFrame resultante
print(data_final)
```
